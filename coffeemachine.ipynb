```python
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}
```

```python
class Ingredient:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity

    def use(self, amount):
        if self.quantity >= amount:
            self.quantity -= amount
            return True
        else:
            return False

    def refill(self, amount):
        self.quantity += amount

```

```python
class Drink:
    def __init__(self, name, price, ingredients_needed):
        self.name = name
        self.price = price
        self.ingredients_needed = ingredients_needed

```

```python
class CoffeeMachine:
    def __init__(self):
        self.ingredients = {
            'water': Ingredient('water', 500),
            'milk': Ingredient('milk', 500),
            'coffee': Ingredient('coffee', 200)
        }
        self.money = 0
        self.drinks = {
            'espresso': Drink('Espresso', 1.5, {'water': 50, 'coffee': 18}),
            'latte': Drink('Latte', 2.5, {'water': 200, 'milk': 150, 'coffee': 24}),
            'cappuccino': Drink('Cappuccino', 3.0, {'water': 250, 'milk': 100, 'coffee': 24})
        }

    def report(self):
        print("Current resources:")
        for name, ingredient in self.ingredients.items():
            print(f"{name.capitalize()}: {ingredient.quantity}ml")
        print(f"Money: ${self.money:.2f}")

    def check_ingredients(self, drink):
        for ingredient, amount_needed in drink.ingredients_needed.items():
            if self.ingredients[ingredient].quantity < amount_needed:
                print(f"Sorry there is not enough {ingredient}.")
                return False
        return True

    def process_coins(self):
        print("Please insert coins.")
        quarters = int(input("How many quarters? ")) * 0.25
        dimes = int(input("How many dimes? ")) * 0.10
        nickles = int(input("How many nickles? ")) * 0.05
        pennies = int(input("How many pennies? ")) * 0.01
        total = quarters + dimes + nickles + pennies
        return round(total, 2)

    def check_transaction(self, money_received, drink_price):
        if money_received < drink_price:
            print("Sorry that's not enough money. Money refunded.")
            return False
        elif money_received > drink_price:
            change = round(money_received - drink_price, 2)
            print(f"Here is ${change} in change.")
        self.money += drink_price
        return True

    def make_drink(self, drink):
        for ingredient, amount_needed in drink.ingredients_needed.items():
            self.ingredients[ingredient].use(amount_needed)
        print(f"Here is your {drink.name}. Enjoy!")

    def start(self):
        machine_on = True
        while machine_on:
            choice = input("What would you like? (espresso/latte/cappuccino): ").lower()
            if choice == "off":
                machine_on = False
                print("Turning off the coffee machine.")
            elif choice == "report":
                self.report()
            elif choice in self.drinks:
                drink = self.drinks[choice]
                if self.check_ingredients(drink):
                    money_received = self.process_coins()
                    if self.check_transaction(money_received, drink.price):
                        self.make_drink(drink)
            else:
                print("Invalid option. Please select espresso, latte, cappuccino, report, or off.")
```

```python

# Makineyi başlatma
coffee_machine = CoffeeMachine()
coffee_machine.start()
```
